# Learning visualization

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Histplot

```python
df = pd.DataFrame({'Scores': [72, 75, 78, 85, 88, 90, 92, 94, 95, 97, 98, 99, 100]})
# using matplotlib
plt.hist(df["Scores"], bins=5, color="green", width=3.2)
plt.xlabel("Scores")
plt.ylabel("Freq of Score range")
plt.title("Hist Diagram for Score")
plt.show()
```

```python
## using sns
sns.histplot(df, x="Scores", bins=5, color="green", kde=True)
plt.xlabel("Score")
plt.ylabel("Freq of score")
plt.title("Histplot using sns")
```
