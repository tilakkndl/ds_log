## Data Distribution

```python
def create_dataframe(n, dist=""):
    """
    Create a DataFrame with n rows and a specified distribution.
    """
    if dist == "uniform":
        data = np.random.uniform(0, 1, n)
    elif dist == "normal":
        data = np.random.normal(0, 1, n)
    elif dist == "exponential":
        data = np.random.exponential(1, n)
    elif dist == "poisson":
        data = np.random.poisson(5, n)
    elif dist=="lognormal":
        data = np.random.lognormal(0, 1, n)
    
    
    df = pd.DataFrame({
        "index": np.arange(n),
        "data": data
    })
    
    return df
# data = np.random.uniform(0, 1, 1000)
# df = pd.DataFrame({
#     "index": np.arange(1000),
#     "data": data
# })
create_dataframe(1000, "uniform")
create_dataframe(1000, "poisson")
create_dataframe(1000, "normal")
create_dataframe(1000, "exponential")
df = create_dataframe(1000, "uniform")
```

```python
sns.histplot(df["data"])
```

```python
plt.scatter(df.index, df.data)
```

```python
df=create_dataframe(1000, "normal")
sns.histplot(df["data"], kde=True)
```

```python
df=create_dataframe(1000, "exponential")
sns.histplot(df["data"], kde=True)
```
