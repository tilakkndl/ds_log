## Correlation
- Ranges from -1 t0 1
- -1 denotes negative correlation
- 0 means there is no relation betweeen two variable
- when values is +1 , it denotes positive correlation

```python
corr_mat=df.drop("species", axis=1).corr()
```

```python
sns.heatmap(corr_mat, annot=True, cmap="coolwarm")
```
