```python
import random
random.seed(42)
idx=random.choices(df.index,k=10)
idx
```

```python
import numpy as np
```

```python
df["Number of employees"][idx]
```

```python
arr=df["Number of employees"].to_numpy().astype("float")
arr, np.isnan(arr).sum()
```

```python
arr[idx]=np.nan
```

```python
np.isnan(arr).sum()
```

```python
df["Number of employees"]=arr
```

```python
df.isna().sum()
```

```python
# mean, median
mean_value=df["Number of employees"].mean()
median_value=df["Number of employees"].median()
```

```python
df['Number of employees']=df['Number of employees'].fillna(mean_value)
```
