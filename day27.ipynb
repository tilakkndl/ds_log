# Intro to Seaborn

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
df
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x="total_bill", y="tip", data=df)
plt.show()
```

- The value of tip is increasing with total bill
- It is seem that   most of the bill lies in the range of 10 and 35

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x="size", y="total_bill", data=df)
plt.show()
```

- when number of people increases up to 4, then total bill also increases .
- When number of people is 5 or 6   , they tend to spend less.
- Discount campaign for 1, 5 or 6 people is effective 

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x="size", y="tip", data=df)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x="day", y="tip", data=df)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x="time", y="tip", data=df)
plt.show()
```

```python
plt.figure(figsize=(4,3))
sns.boxplot(x="day", y="total_bill", data=df)
plt.show()
```

```python
plt.figure(figsize=(4,3))
sns.violinplot(x="day", y="total_bill", data=df)
plt.show()
```

```python
plt.figure(figsize=(4,3))
sns.histplot(df["total_bill"], bins=20, kde=True)
plt.show()
```

kde=kernel density e

```python
plt.figure(figsize=(4,3))
sns.kdeplot(df["total_bill"], shade=True)
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```
