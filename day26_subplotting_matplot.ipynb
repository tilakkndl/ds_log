# Subplotting and multiploting using matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec # to be reviews
```

```python
x=np.arange(0,100,10)
x
```

```python
np.sin(x), np.cos(x)
```

### .subplot()

```python
plt.figure(figsize=(6,3))
plt.plot(x)
# plt.plot(np.sin(x))
plt.show()
```

```python
plt.plot(np.sin(x))
```

```python
plt.subplot(2,2,1)
plt.scatter(np.arange(len(x)), x)
plt.xlabel('NOf of x')
plt.ylabel('Value of x')
plt.title('Scatter plot of x')

plt.subplot(2,2,2)
plt.plot(x)
plt.ylabel("x")

plt.subplot(2,2,3)

plt.subplot(2,2,4)
plt.scatter(np.arange(len(x)), np.exp(x))
plt.xlabel('NOf of x')
plt.ylabel('Value of exp(x)')
plt.title('Scatter plot of exp(x)')

plt.suptitle('Subplots of x')
plt.tight_layout()
plt.show()
```

### using .subplots()

```python
fig, ax=plt.subplots(3,2, sharex=True, sharey=True)
fig, ax
print(fig)
ax[0,0].scatter(np.arange(len(x)), x)
ax[0,0].set_title('Scatter plot of x')

ax[0,0].scatter(np.arange(len(x)), np.cos(x))
ax[0,0].set_title('Scatter plot of x')

ax[1,1].plot(np.sin(x))
ax[1,1].set_title('Plot of sin(x)')
ax[1,1].set_xlabel('x')
ax[1,1].set_ylabel('sin(x)')

ax[2,0].plot(np.exp(x))
ax[2,0].set_title('Plot of exp(x)')
ax[2,0].set_xlabel('x')
ax[2,0].set_ylabel('exp(x)')

plt.suptitle('Subplots of x')
plt.tight_layout()
```

### subploting using gridspec

```python
fig=plt.figure()
gs=gridspec.GridSpec(2,2)
ax1=fig.add_subplot(gs[0,0])
ax2=fig.add_subplot(gs[0,1])
ax3=fig.add_subplot(gs[1,:])
ax1.scatter(np.arange(len(x)), x)
ax1.set_title('Scatter plot of x')  
ax2.scatter(np.arange(len(x)), np.cos(x))
ax2.set_title('Scatter plot of cos(x)')
ax3.plot(np.sin(x))
ax3.set_title('Plot of sin(x)')
ax3.set_xlabel('x')
ax3.set_ylabel('sin(x)')
plt.tight_layout()
plt.suptitle('Subplots of x')
```
